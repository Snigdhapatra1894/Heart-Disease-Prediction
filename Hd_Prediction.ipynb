```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
heart_disease = pd.read_csv("https://raw.githubusercontent.com/Snigdhapatra1894/Heart-Disease-Prediction/main/heart.csv")
heart_disease
```

```python
heart_disease.head()
```

```python
heart_disease.info()
```

```python
heart_disease.isnull().sum().sum()
```

```python
# no NaN value found .
fig, ax = plt.subplots(figsize=(15, 15))
correl = heart_disease.corr()
ax = sns.heatmap(correl,
                 annot=True,
                 linewidths=0.3,
                 fmt=".2f",
                 cmap="mako" ,
                 cbar=True);
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 1, top - 0.5) ;
```

```python
x = heart_disease.drop("target", axis=1)
y = heart_disease["target"]
np.random.seed(1894)
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x ,y , test_size=0.15)
len(x_train) , len(y_train) , len(x_test) , len(y_test)
```

```python
#Decision Trees
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(x_train,y_train)
predictions = dtree.predict(x_test)
```
